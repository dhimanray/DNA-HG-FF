```python
import numpy as np
import matplotlib.pyplot as plt
import uncertainties as unc  
import uncertainties.unumpy as unumpy
```

```python
all_entropy = []

ch27wc = []
for i in range(5):
    e1 = np.loadtxt('CHARMM27/WC_trajectories/entropy_%d/entropy.dat'%(i+1))
    e1 = e1*298
    ch27wc.append(e1)
ch27wc = np.asarray(ch27wc)
all_entropy.append([np.mean(ch27wc),1.243*np.std(ch27wc)])

ch27hg = []
for i in range(5):
    e1 = np.loadtxt('CHARMM27/HG_trajectories/entropy_%d/entropy.dat'%(i+1))
    e1 = e1*298
    ch27hg.append(e1)
ch27hg = np.asarray(ch27hg)
all_entropy.append([np.mean(ch27hg),1.243*np.std(ch27hg)])

ch36wc = []
for i in range(5):
    e1 = np.loadtxt('CHARMM36/WC_trajectories/entropy_%d/entropy.dat'%(i+1))
    e1 = e1*298
    ch36wc.append(e1)
ch36wc = np.asarray(ch36wc)
all_entropy.append([np.mean(ch36wc),1.243*np.std(ch36wc)])

ch36hg = []
for i in range(5):
    e1 = np.loadtxt('CHARMM36/HG_trajectories/entropy_%d/entropy.dat'%(i+1))
    e1 = e1*298
    ch36hg.append(e1)
ch36hg = np.asarray(ch36hg)
all_entropy.append([np.mean(ch36hg),1.243*np.std(ch36hg)])

amber_bsc0_wc = []
for i in range(5):
    e1 = np.loadtxt('AMBER_bsc_0/WC_trajectories/entropy_%d/entropy.dat'%(i+1))
    e1 = e1*298
    amber_bsc0_wc.append(e1)
amber_bsc0_wc = np.asarray(amber_bsc0_wc)
all_entropy.append([np.mean(amber_bsc0_wc),1.243*np.std(amber_bsc0_wc)])

amber_bsc0_hg = []
for i in range(5):
    e1 = np.loadtxt('AMBER_bsc_0/HG_trajectories/entropy_%d/entropy.dat'%(i+1))
    e1 = e1*298
    amber_bsc0_hg.append(e1)
amber_bsc0_hg = np.asarray(amber_bsc0_hg)
all_entropy.append([np.mean(amber_bsc0_hg),1.243*np.std(amber_bsc0_hg)])

amber_bsc1_wc = []
for i in range(5):
    e1 = np.loadtxt('AMBER_bsc_1/WC_trajectories/entropy_%d/entropy.dat'%(i+1))
    e1 = e1*298
    amber_bsc1_wc.append(e1)
amber_bsc1_wc = np.asarray(amber_bsc1_wc)
all_entropy.append([np.mean(amber_bsc1_wc),1.243*np.std(amber_bsc1_wc)])

amber_bsc1_hg = []
for i in range(5):
    e1 = np.loadtxt('AMBER_bsc_1/HG_trajectories/entropy_%d/entropy.dat'%(i+1))
    e1 = e1*298
    amber_bsc1_hg.append(e1)
amber_bsc1_hg = np.asarray(amber_bsc1_hg)
all_entropy.append([np.mean(amber_bsc1_hg),1.243*np.std(amber_bsc1_hg)])

all_entropy = np.array(all_entropy)
```

```python
print(all_entropy)
```

```python
forceField = ['CHARMM27_WC','CHARMM27_HG','CHARMM36_WC','CHARMM36_HG','AMBER_bsc_0_WC','AMBER_bsc_0_HG','AMBER_bsc_1_WC','AMBER_bsc_1_HG']
entropy = all_entropy.T[0]
error_bars = all_entropy.T[1]
```

```python
colors = ['b','r','b','r','b','r','b','r']

fig = plt.figure(figsize=(12,5))
ax = fig.add_axes([0,0,1,1])
ax.bar(forceField,entropy,yerr=error_bars,capsize=20.0,color=colors)
plt.ylim(14)
plt.yticks(fontsize=16)
plt.xticks(color='white')
plt.ylabel('T$\Delta$S (kcal/mol)',fontsize=20)
plt.text(-0.1,15,'CHARMM27 (WC)',fontsize=20,rotation='vertical',color='white')
plt.text(0.9,15,'CHARMM27 (HG)',fontsize=20,rotation='vertical',color='white')
plt.text(1.9,15,'CHARMM36 (WC)',fontsize=20,rotation='vertical',color='white')
plt.text(2.9,15,'CHARMM36 (HG)',fontsize=20,rotation='vertical',color='white')
plt.text(3.9,14.5,'AMBERbsc0 (WC)',fontsize=18,rotation='vertical',color='white')
plt.text(4.9,14.5,'AMBERbsc0 (HG)',fontsize=16,rotation='vertical',color='white')
plt.text(5.9,14.5,'AMBERbsc1 (WC)',fontsize=18,rotation='vertical',color='white')
plt.text(6.9,14.5,'AMBERbsc1 (HG)',fontsize=16,rotation='vertical',color='white')
#plt.tight_layout()
#plt.show()
plt.savefig('absolute_entropy.png',dpi=300,bbox_inches='tight')
```

```python
en_wc = np.asarray([all_entropy[0,0],all_entropy[2,0],all_entropy[4,0],all_entropy[6,0]])
en_wc_err = np.asarray([all_entropy[0,1],all_entropy[2,1],all_entropy[4,1],all_entropy[6,1]])
en_hg = np.asarray([all_entropy[1,0],all_entropy[3,0],all_entropy[5,0],all_entropy[7,0]])
en_hg_err = np.asarray([all_entropy[1,1],all_entropy[3,1],all_entropy[5,1],all_entropy[7,1]])

wc = unumpy.uarray(( en_wc, en_wc_err ))
hg = unumpy.uarray(( en_hg, en_hg_err ))

diff = hg - wc

mean_diff = unumpy.nominal_values(diff)
err_diff = unumpy.std_devs(diff)
```

```python
#difference = [e2-e1, e4-e3, e6-e5, e8-e7]
forceFF = ['CHARMM27','CHARMM36','AMBER_bsc_0','AMBER_bsc_1']

fig = plt.figure(figsize=(9,4))
ax = fig.add_axes([0,0,1,1])
ax.bar(forceFF,mean_diff,yerr=err_diff,capsize=20,color='lightgreen')
#plt.ylim(15)
plt.yticks(fontsize=14)
plt.ylabel('T$\Delta$S (HG) - T$\Delta$S (WC) (kcal/mol)',fontsize=16)
plt.xticks(fontsize=14)
plt.axhline(y=0.0)
#plt.tight_layout()
#plt.show()
plt.savefig('difference_entropy.png',dpi=300,bbox_inches='tight')
```

```python
ch27wc_rmsf = []
for i in range(5):
    l = np.loadtxt('CHARMM27/WC_trajectories/entropy_%d/rmsf.dat'%(i+1))
    x = l[:,0]
    y = l[:,1]
    ch27wc_rmsf.append(y)
ch27wc_rmsf = np.array(ch27wc_rmsf)
wc_mean = np.mean(ch27wc_rmsf,axis=0)
wc_err = 1.243*np.std(ch27wc_rmsf,axis=0)

ch27hg_rmsf = []
for i in range(5):
    l = np.loadtxt('CHARMM27/HG_trajectories/entropy_%d/rmsf.dat'%(i+1))
    x = l[:,0]
    y = l[:,1]
    ch27hg_rmsf.append(y)
ch27hg_rmsf = np.array(ch27hg_rmsf)
hg_mean = np.mean(ch27hg_rmsf,axis=0)
hg_err = 1.243*np.std(ch27hg_rmsf,axis=0)

wc = unumpy.uarray(( wc_mean, wc_err ))
hg = unumpy.uarray(( hg_mean, hg_err ))

ratio = hg/wc

mean_ratio = unumpy.nominal_values(ratio)
err_ratio = unumpy.std_devs(ratio)

#----------- Plotting -----------------------------
fig = plt.figure(figsize=(15,5))
plt.plot(x,mean_ratio,'g')
plt.fill_between(x, mean_ratio - err_ratio, mean_ratio + err_ratio, facecolor='xkcd:lightgreen')
plt.title('CHARMM27',fontsize=16)
#plt.axvline(x=[254,286,474,506])
plt.axvline(x=254,color='k')
plt.axvline(x=286,color='k')
plt.axvline(x=474,color='k')
plt.axvline(x=506,color='k')
plt.axhline(y=1.0,color='b')
plt.xlabel('Heavy atom index',fontsize=16)
plt.ylabel('RMSF Ratio (HG/WC) ',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(0,763)
plt.text(260,1.4,'T9',fontsize=16)
plt.text(475,1.4,'A16',fontsize=16)
#plt.show()
plt.savefig('CHARMM27_rmsf.png',dpi=300,bbox_inches='tight')


charmm27_mean_rmsf_ratio = mean_ratio
charmm27_err_rmsf_ratio = err_ratio
```
